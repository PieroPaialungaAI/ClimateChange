```python
X=np.array(la.AverageTemperature)
```

```python
X=X.reshape(-1,1)
```

```python
kde = KernelDensity(kernel='gaussian', bandwidth=0.5).fit(X)
```

```python
MAX_EXTR=[]
for i in range(5000):
    MAX_EXTR.append(kde.sample(12).max())
```

```python
my_kde = sns.kdeplot(MAX_EXTR,bw=0.5,label='Kernel Density Estimator',color='k')
line = my_kde.lines[0]
x, y = line.get_data()
sns.distplot(MAX_EXTR,norm_hist=True,kde=False,label='Data Histogram')
#plt.plot(sns.distplot(kde.sample(1000)))
plt.grid(True)
plt.legend()
```

```python
NEW_EXTR=np.array(MAX_EXTR).reshape(-1,1)
kde = KernelDensity(kernel='gaussian', bandwidth=0.2).fit(NEW_EXTR)
```

```python
data_sample=kde.sample(1000).reshape(1,-1)[0]
```
